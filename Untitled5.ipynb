```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def stakeholder_simulation(model_type, time_steps=50, stakeholder_investment_strategy=None):
    # Initialize metrics
    profit = [1.0]
    employee_satisfaction = [0.5]
    customer_satisfaction = [0.5]
    community_impact = [0.5]
    supplier_relations = [0.5]  # New stakeholder
    sustainability_index = [0.5]
    stakeholder_value = [1.0 + 0.3*0.5 + 0.3*0.5 + 0.2*0.5 + 0.2*0.5]  # Initial stakeholder value
    scenarios = ["None"]  # Initialize with initial state

    # Parameters
    growth_rate = 0.02
    customer_impact_factor = 0.15
    employee_productivity_factor = 0.1
    supplier_stability_factor = 0.05
    employee_investment_factor = 0.08
    customer_investment_factor = 0.07
    community_investment_factor = 0.05
    supplier_investment_factor = 0.06
    employee_dissatisfaction_decay = 0.01
    customer_dissatisfaction_decay = 0.015
    community_degradation_decay = 0.005
    supplier_instability_decay = 0.005
    stakeholder_weights = {'employee': 0.3, 'customer': 0.3, 'community': 0.2, 'supplier': 0.2}

    np.random.seed(42)  # For reproducibility

    for t in range(time_steps):
        investment = 0
        if model_type == 'stakeholder' and stakeholder_investment_strategy:
            investment = stakeholder_investment_strategy(t, profit[-1], employee_satisfaction[-1],
                                                        customer_satisfaction[-1], community_impact[-1],
                                                        supplier_relations[-1])

        # Copy current values to modify in scenarios
        current_profit = profit[-1]
        current_customer_satisfaction = customer_satisfaction[-1]
        current_employee_satisfaction = employee_satisfaction[-1]
        current_community_impact = community_impact[-1]
        current_supplier_relations = supplier_relations[-1]
        current_sustainability_index = sustainability_index[-1]
        scenario = "None"

        # Scenario 1: Reputation Crisis
        if current_customer_satisfaction < 0.4 and t > 10 and np.random.random() < 0.3:
            profit_delta = np.random.uniform(-0.2, -0.1)
            current_profit += profit_delta
            current_customer_satisfaction -= 0.05
            scenario = "Reputation Crisis"

        # Scenario 2: Regulatory Change
        elif t == 20 and np.random.random() < 0.25:
            profit_delta = np.random.uniform(-0.15, -0.05)
            current_profit += profit_delta
            current_community_impact += np.random.uniform(0.05, 0.1)
            scenario = "Regulatory Change"

        # Scenario 3: Employee Strike
        elif current_employee_satisfaction < 0.35 and t > 15 and np.random.random() < 0.2:
            profit_delta = np.random.uniform(-0.15, -0.05)
            current_profit += profit_delta
            current_customer_satisfaction += profit_delta * 0.5
            current_employee_satisfaction -= 0.05
            scenario = "Employee Strike"

        # Scenario 4: Supply Chain Disruption
        elif t == 30 and np.random.random() < 0.2:
            profit_delta = np.random.uniform(-0.2, -0.1)
            current_supplier_relations += profit_delta
            current_profit += profit_delta * 0.7
            current_customer_satisfaction += profit_delta * 0.3
            scenario = "Supply Chain Disruption"

        # Scenario 5: Sustainability Initiative Success
        elif current_community_impact > 0.7 and t > 25 and np.random.random() < 0.25:
            positive_impact = np.random.uniform(0.05, 0.15)
            current_customer_satisfaction += positive_impact
            current_profit += positive_impact * 0.5
            current_sustainability_index += positive_impact
            scenario = "Sustainability Success"

        # Update metrics
        next_profit = current_profit * (1 + growth_rate) + \
                      (current_customer_satisfaction * customer_impact_factor) + \
                      (current_employee_satisfaction * employee_productivity_factor) + \
                      (current_supplier_relations * supplier_stability_factor) - investment
        next_employee_satisfaction = current_employee_satisfaction + \
                                    (investment * employee_investment_factor) - \
                                    (employee_dissatisfaction_decay * (1 - current_employee_satisfaction))
        next_customer_satisfaction = current_customer_satisfaction + \
                                    (investment * customer_investment_factor) + \
                                    (current_employee_satisfaction * 0.05) - \
                                    (customer_dissatisfaction_decay * (1 - current_customer_satisfaction))
        next_community_impact = current_community_impact + \
                                (investment * community_investment_factor) - \
                                (community_degradation_decay * (1 - current_community_impact))
        next_supplier_relations = current_supplier_relations + \
                                  (investment * supplier_investment_factor) - \
                                  (supplier_instability_decay * (1 - current_supplier_relations))
        next_sustainability_index = (0.4 * next_community_impact + 0.3 * next_employee_satisfaction +
                                    0.3 * next_customer_satisfaction)

        # Append new values
        profit.append(max(0, next_profit))
        employee_satisfaction.append(max(0, min(1, next_employee_satisfaction)))
        customer_satisfaction.append(max(0, min(1, next_customer_satisfaction)))
        community_impact.append(max(0, min(1, next_community_impact)))
        supplier_relations.append(max(0, min(1, next_supplier_relations)))
        sustainability_index.append(max(0, min(1, next_sustainability_index)))

        sv = profit[-1] + \
             sum(stakeholder_weights[k] * v[-1] for k, v in
                 [('employee', employee_satisfaction), ('customer', customer_satisfaction),
                  ('community', community_impact), ('supplier', supplier_relations)])
        stakeholder_value.append(sv)
        scenarios.append(scenario)

    # Verify lengths
    lengths = [len(profit), len(employee_satisfaction), len(customer_satisfaction),
               len(community_impact), len(supplier_relations), len(sustainability_index),
               len(stakeholder_value), len(scenarios)]
    if len(set(lengths)) != 1:
        raise ValueError(f"Array lengths mismatch: {lengths}")

    return {
        'profit': profit,
        'employee_satisfaction': employee_satisfaction,
        'customer_satisfaction': customer_satisfaction,
        'community_impact': community_impact,
        'supplier_relations': supplier_relations,
        'sustainability_index': sustainability_index,
        'stakeholder_value': stakeholder_value,
        'scenarios': scenarios
    }

def balanced_investment_strategy(t, profit, es, cs, ci, sr):
    return profit * 0.05

def reactive_investment_strategy(t, profit, es, cs, ci, sr):
    threshold = 0.6
    if any(x < threshold for x in [es, cs, ci, sr]):
        return profit * 0.10
    return profit * 0.02

# Run simulations
time_steps = 50
shareholder_results = stakeholder_simulation('shareholder', time_steps)
stakeholder_balanced_results = stakeholder_simulation('stakeholder', time_steps,
                                                     stakeholder_investment_strategy=balanced_investment_strategy)
stakeholder_reactive_results = stakeholder_simulation('stakeholder', time_steps,
                                                     stakeholder_investment_strategy=reactive_investment_strategy)

# Prepare DataFrame for plotting
def prepare_dataframe(results, model_name):
    return pd.DataFrame({
        'Time': range(len(results['profit'])),
        'Profit': results['profit'],
        'Stakeholder Value': results['stakeholder_value'],
        'Employee Satisfaction': results['employee_satisfaction'],
        'Customer Satisfaction': results['customer_satisfaction'],
        'Community Impact': results['community_impact'],
        'Supplier Relations': results['supplier_relations'],
        'Sustainability Index': results['sustainability_index'],
        'Model': model_name,
        'Scenario': results['scenarios']
    })

df_shareholder = prepare_dataframe(shareholder_results, 'Shareholder')
df_balanced = prepare_dataframe(stakeholder_balanced_results, 'Stakeholder (Balanced)')
df_reactive = prepare_dataframe(stakeholder_reactive_results, 'Stakeholder (Reactive)')
combined_df = pd.concat([df_shareholder, df_balanced, df_reactive])

# Set seaborn style
sns.set(style="whitegrid", palette="deep")

# Plotting
plt.figure(figsize=(16, 12))

# Plot 1: Profit
plt.subplot(2, 2, 1)
sns.lineplot(data=combined_df, x='Time', y='Profit', hue='Model',
             palette={'Shareholder': 'red', 'Stakeholder (Balanced)': 'green', 'Stakeholder (Reactive)': 'blue'})
plt.title('Profit Over Time')
plt.xlabel('Time Step')
plt.ylabel('Profit')

# Plot 2: Stakeholder Value
plt.subplot(2, 2, 2)
sns.lineplot(data=combined_df, x='Time', y='Stakeholder Value', hue='Model',
             palette={'Shareholder': 'red', 'Stakeholder (Balanced)': 'green', 'Stakeholder (Reactive)': 'blue'})
plt.title('Stakeholder Value Over Time')
plt.xlabel('Time Step')
plt.ylabel('Stakeholder Value')

# Plot 3: Stakeholder Metrics
plt.subplot(2, 2, 3)
for metric, color in [('Employee Satisfaction', 'purple'), ('Customer Satisfaction', 'orange'),
                      ('Community Impact', 'brown'), ('Supplier Relations', 'cyan')]:
    for model, style in [('Shareholder', '--'), ('Stakeholder (Balanced)', '-'), ('Stakeholder (Reactive)', '-.')]:
        sns.lineplot(data=combined_df[combined_df['Model'] == model], x='Time', y=metric,
                     label=f'{model} {metric}', linestyle=style, color=color, alpha=0.6)
plt.title('Stakeholder Metrics')
plt.xlabel('Time Step')
plt.ylabel('Level (0-1)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot 4: Sustainability Index
plt.subplot(2, 2, 4)
sns.lineplot(data=combined_df, x='Time', y='Sustainability Index', hue='Model',
             palette={'Shareholder': 'red', 'Stakeholder (Balanced)': 'green', 'Stakeholder (Reactive)': 'blue'})
plt.title('Sustainability Index Over Time')
plt.xlabel('Time Step')
plt.ylabel('Sustainability Index (0-1)')

plt.tight_layout()
plt.savefig('stakeholder_simulation.png')

# Numerical Results
def calculate_cagr(values, time_steps):
    if values[0] <= 0 or values[-1] <= 0 or time_steps <= 0:
        return 0
    return (np.power(values[-1] / values[0], 1 / time_steps) - 1) * 100

print("\n=== Numerical Results ===")
for name, results in [('Shareholder', shareholder_results),
                      ('Stakeholder (Balanced)', stakeholder_balanced_results),
                      ('Stakeholder (Reactive)', stakeholder_reactive_results)]:
    print(f"\n{name}:")
    print(f"  Profit CAGR: {calculate_cagr(results['profit'], time_steps):.2f}%")
    print(f"  Stakeholder Value CAGR: {calculate_cagr(results['stakeholder_value'], time_steps):.2f}%")
    print(f"  Final Employee Satisfaction: {results['employee_satisfaction'][-1]:.2f}")
    print(f"  Final Customer Satisfaction: {results['customer_satisfaction'][-1]:.2f}")
    print(f"  Final Community Impact: {results['community_impact'][-1]:.2f}")
    print(f"  Final Supplier Relations: {results['supplier_relations'][-1]:.2f}")
    print(f"  Final Sustainability Index: {results['sustainability_index'][-1]:.2f}")

print("\n=== Scenario Events ===")
for name, results in [('Shareholder', shareholder_results),
                      ('Stakeholder (Balanced)', stakeholder_balanced_results),
                      ('Stakeholder (Reactive)', stakeholder_reactive_results)]:
    print(f"\n{name} Scenarios:")
    for t, scenario in enumerate(results['scenarios']):
        if scenario != "None":
            print(f"  Time Step {t}: {scenario}")
```
